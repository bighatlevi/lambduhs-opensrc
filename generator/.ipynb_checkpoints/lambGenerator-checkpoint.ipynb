```python
# Lamb Duhs Generator
# @dev: bighatlevi

# Special thanks to vortextemporum and the Boring Bananas Company!

from PIL import Image, ImageOps
from IPython.display import display
import random
import json


## Define asset names and accessory points

backgroundColors = ["Epsilon", "Delta", "Gamma", "Beta", "Alpha", "Omega"]
woolColors = ["Black", "Blue", "Brown", "Cyan", "Gray", "Green", "Orange", "Pink", "Purple", "Red", "White", "Yellow"]

faces = [
    "Autopilot",
    "Big Smile", 
    "Disappointed", 
    "Duh", 
    "Dumbfounded", 
    "Eyepatch", 
    "Flustered", 
    "Galaxy", 
    "Glasses", 
    "Happy", 
    "Heartglasses", 
    "Hungry", 
    "Jim",
    "Okay",
    "Pixelated", 
    "Pog", 
    "Pout", 
    "Slant", 
    "Sleepy", 
    "Smile", 
    "Smirk", 
    "Smug", 
    "Stuffed", 
    "Sunglasses", 
    "Tongue", 
    "Warrior",
]

shirts = {
    "Astro Suit" : 4,
    "Baaka Shirt" : 1,
    "Baasa Shirt" : 1,
    "Baatalica Shirt" : 1,
    "Candy Hoodie" : 2,
    "Chops Shirt" : 1,
    "Corduroy Jacket" : 2,
    "Cowboy Vest" : 3,
    "Dress" : 2,
    "Flannel" : 2,
    "Flower Lei" : 1,
    "Kevlar Vest" : 1,
    "Lab Coat" : 3,
    "Leather Jacket" : 2,
    "Overalls" : 1,
    "Parka" : 3,
    "Peacoat" : 3,
    "Royal Cape" : 4,
    "Rune Robe" : 2,
    "Sage Robe" : 3,
    "Scarf" : 1,
    "Scuba Suit" : 1,
    "Spartan Armor" : 4,
    "Sports Jersey" : 1,
    "Sweater" : 2,
    "Taco Shirt" : 1,
    "Turtleneck" : 2,
    "Tuxedo" : 3,
    "Wet Shirt" : 1,
    "Zip Hoodie" : 2,
}

hats = {
    "Afro" : 2,
    "Antennas" : 1,
    "Astro Helmet" : 4,
    "Ball Cap" : 1,
    "Beanie" : 3,
    "Beret" : 1,
    "Big Horns" : 3,
    "Bucket Hat" : 1,
    "Cowboy Hat" : 2,
    "Crown" : 3,
    "Earbuds" : 1,
    "Fedora" : 1,
    "Fire Hat" : 3,
    "Fish Cap" : 2,
    "Flower" : 1,
    "Gamer Headset" : 2,
    "Hard Hat" : 2,
    "Headband" : 1,
    "Head Mirror" : 1,
    "Jester Hat" : 3,
    "Little Horns" : 1,
    "Medium Horns" : 2,
    "Military Helmet" : 2,
    "Night Goggles": 3,
    "Party Hat" : 1,
    "Pilot Hat" : 3,
    "Pineapple" : 1,
    "Pirate Hat" : 2,
    "Police Hat" : 2,
    "Sage Hat" : 4,
    "Scuba Goggles" : 1,
    "Sideways Cap" : 1,
    "Skull Clip" : 1,
    "Spartan Helmet" : 4,
    "Straw Hat" : 2,
    "Tiara" : 3,
    "Top Hat" : 4,
    "Ushanka" : 2,
    "VR Headset" : 2,
    "Wow Cap" : 1
}


## Define asset files
backgroundFiles = {
    "Epsilon" : "bg_yellow",
    "Delta" : "bg_green",
    "Gamma" : "bg_blue",
    "Beta" : "bg_purple",
    "Alpha" : "bg_red",
    "Omega" : "bg_orange",
}

woolFiles = {
    "Black" : "black",
    "Blue" : "blue",
    "Brown" : "brown",
    "Cyan" : "cyan",
    "Gray" : "gray",
    "Green" : "green",
    "Orange" : "orange",
    "Pink" : "pink",
    "Purple" : "purple",
    "Red" : "red",
    "White" : "white",
    "Yellow" : "yellow",
}

faceFiles = {
    "Autopilot" : "autopilot",
    "Big Smile" : "bigsmile",
    "Disappointed" : "disappointed",
    "Duh" : "duh",
    "Dumbfounded" : "dumbfounded",
    "Eyepatch" : "eyepatch",
    "Flustered" : "flustered",
    "Galaxy" : "galaxy",
    "Glasses" : "glasses",
    "Happy" : "happy",
    "Heartglasses" : "heartglasses",
    "Hungry" : "hungry",
    "Jim" : "jim",
    "Okay" : "okay",
    "Pixelated" : "pixelated",
    "Pog" : "pog",
    "Pout" : "pout",
    "Slant" : "slant",
    "Sleepy" : "sleepy",
    "Smile" : "smile",
    "Smirk" : "smirk",
    "Smug" : "smug",
    "Stuffed" : "stuffed",
    "Sunglasses" : "sunglasses",
    "Tongue" : "tongue",
    "Warrior" : "warrior",
}

shirtFiles = {
    "Astro Suit" : "astrosuit",
    "Baaka Shirt" : "baaka",
    "Baasa Shirt" : "baasa",
    "Baatalica Shirt" : "baatalica",
    "Candy Hoodie" : "candyhoodie",
    "Chops Shirt" : "chops",
    "Corduroy Jacket" : "corduroy",
    "Cowboy Vest" : "cowboyvest",
    "Dress" : "dress",
    "Flannel" : "flannel",
    "Flower Lei" : "flowerlei",
    "Kevlar Vest" : "kevlar",
    "Lab Coat" : "labcoat",
    "Leather Jacket" : "leather",
    "Overalls" : "overalls",
    "Parka" : "parka",
    "Peacoat" : "peacoat",
    "Royal Cape" : "royalcape",
    "Rune Robe" : "runerobe",
    "Sage Robe" : "sagerobe",
    "Scarf" : "scarf",
    "Scuba Suit" : "scubasuit",
    "Spartan Armor": "spartanarmor",
    "Sports Jersey" : "sportsjersey",
    "Sweater" : "sweater",
    "Taco Shirt" : "taco",
    "Turtleneck" : "turtleneck",
    "Tuxedo" : "tuxedo",
    "Wet Shirt" : "wetshirt",
    "Zip Hoodie" : "ziphoodie"
}

hatFiles = {
    "Afro" : "afro",
    "Antennas" : "antennas",
    "Astro Helmet" : "astrohelm",
    "Ball Cap" : "ballcap",
    "Beanie" : "beanie",
    "Beret" : "beret",
    "Big Horns" : "bighorns",
    "Bucket Hat" : "buckethat",
    "Cowboy Hat" : "cowboyhat",
    "Crown" : "crown",
    "Earbuds" : "earbuds",
    "Fedora" : "fedora",
    "Fire Hat" : "firehat",
    "Fish Cap" : "fishcap",
    "Flower" : "flower",
    "Gamer Headset" : "gamerheadset",
    "Hard Hat" : "hardhat",
    "Headband" : "headband",
    "Head Mirror" : "headmirror",
    "Jester Hat" : "jesterhat",
    "Little Horns" : "lilhorns",
    "Medium Horns" : "medhorns",
    "Military Helmet" : "militaryhelm",
    "Night Goggles" : "nightgoggles",
    "Party Hat" : "partyhat",
    "Pilot Hat" : "pilothat",
    "Pineapple" : "pineapple",
    "Pirate Hat" : "piratehat",
    "Police Hat" : "policehat",
    "Sage Hat" : "sagehat",
    "Scuba Goggles" : "scubagoggles",
    "Sideways Cap" : "sideways",
    "Skull Clip" : "skullclip",
    "Spartan Helmet" : "spartanhelm",
    "Straw Hat" : "strawhat",
    "Tiara" : "tiara",
    "Top Hat" : "tophat",
    "Ushanka" : "ushanka",
    "VR Headset" : "vrheadset",
    "Wow Cap" : "wowcap"
}


## Define functions

# Creates different trait combinations
def createTraitCombo():
    trait = {}

    randomShirt = random.choices(list(shirts.keys()))[0]
    shirtPoints = shirts[randomShirt]
    
    randomHat = random.choices(list(hats.keys()))[0]
    hatPoints = hats[randomHat]
    
    pointTotal = shirtPoints + hatPoints
    background = determineRarity(pointTotal)
    
    trait["Tier"] = background
    trait["Wool"] = random.choices(woolColors)[0]
    trait["Face"] = random.choices(faces)[0]
    trait["Body"] = randomShirt
    trait["Head"] = randomHat
    
    if trait in traits:
        # If current lamb trait combo is already found in trait dictionary, restart function
        return createTraitCombo()
    else:
        return trait

# Creates the custom lambs for the team
def createTeamLambs(index):
    trait = {}
    
    if index == 0:
        trait["Tier"] = "Alpha"
        trait["Wool"] = "Blue" 
        trait["Face"] = "Sunglasses"
        trait["Body"] = "Lab Coat"
        trait["Head"] = "Sage Hat"
        
    elif index == 1:
        trait["Tier"] = "Gamma"
        trait["Wool"] = "Purple" 
        trait["Face"] = "Okay"
        trait["Body"] = "Parka"
        trait["Head"] = "Fish Cap"
        
    elif index == 2:
        trait["Tier"] = "Epsilon"
        trait["Wool"] = "Green" 
        trait["Face"] = "Tongue"
        trait["Body"] = "Scarf"
        trait["Head"] = "Skull Clip"
        
    elif index == 3:
        trait["Tier"] = "Beta"
        trait["Wool"] = "Black" 
        trait["Face"] = "Pixelated"
        trait["Body"] = "Candy Hoodie"
        trait["Head"] = "Spartan Helmet"
        
    elif index == 4:
        trait["Tier"] = "Delta"
        trait["Wool"] = "Red" 
        trait["Face"] = "Hungry"
        trait["Body"] = "Taco Shirt"
        trait["Head"] = "Night Goggles"
        
    elif index == 5:
        trait["Tier"] = "Epsilon"
        trait["Wool"] = "Pink" 
        trait["Face"] = "Heartglasses"
        trait["Body"] = "Flower Lei"
        trait["Head"] = "Tiara"
    return trait

# Determines rarity tier/background color based on calculated point total
def determineRarity(total):
    switch = {
        2 : "Epsilon",
        3 : "Epsilon",    # Yellow
        4 : "Delta",      # Green
        5 : "Gamma",      # Blue
        6 : "Beta",       # Purple
        7 : "Alpha",      # Red
        8 : "Omega"       # Orange
    }
    return switch.get(total)

# Checks if all lambs are unique
def allUnique(attributes):
    seen = list()
    return not any(i in seen or seen.append(i) for i in attributes)


## This is where the magic happens

# Create list of asset combinations
TOTAL_LAMBS = 20
traits = []

for i in range(TOTAL_LAMBS):
    if (i > 5):
        newTraitCombo = createTraitCombo()
        traits.append(newTraitCombo)
    else:
        teamLambCombo = createTeamLambs(i)
        traits.append(teamLambCombo)

if(allUnique(traits) == False):
    print("Not all lambs are unique, regenerate")
else:
    print("All lambs are unique")
    
# Add token IDs to metadata (JSON)
id = 0
for item in traits:
    item["tokenId"] = id
    id = id + 1

# Show completed traits list for testing purposes
# print(traits)

# Get trait counts
backgroundCount = {}
for item in backgroundColors:
    backgroundCount[item] = 0
    
woolCount = {}
for item in woolColors:
    woolCount[item] = 0
    
faceCount = {}
for item in faces:
    faceCount[item] = 0
    
shirtCount = {}
for item in shirts:
    shirtCount[item] = 0
    
hatCount = {}
for item in hats:
    hatCount[item] = 0
    
for lamb in traits:
    #print(lamb)
    backgroundCount[lamb["Tier"]] += 1
    woolCount[lamb["Wool"]] += 1
    faceCount[lamb["Face"]] += 1
    shirtCount[lamb["Body"]] += 1
    hatCount[lamb["Head"]] += 1
    
print(backgroundCount)
print(woolCount)
print(faceCount)
print(shirtCount)
print(hatCount)
```

```python
# Image generation
    
for item in traits:

    img1 = Image.open(f'../../assets/backgrounds/{backgroundFiles[item["Tier"]]}.png').convert('RGBA')
    img2 = Image.open(f'../../assets/wool/{woolFiles[item["Wool"]]}.png').convert('RGBA')
    img3 = Image.open(f'../../assets/wool/outline.png').convert('RGBA')
    img4 = Image.open(f'../../assets/shirts/{shirtFiles[item["Body"]]}.png').convert('RGBA')
    img5 = Image.open(f'../../assets/faces/{faceFiles[item["Face"]]}.png').convert('RGBA')
    img6 = Image.open(f'../../assets/hats/{hatFiles[item["Head"]]}.png').convert('RGBA')

    # Create each composite
    com1 = Image.alpha_composite(img1, img2)
    com2 = Image.alpha_composite(com1, img3)
    com3 = Image.alpha_composite(com2, img4)
    com4 = Image.alpha_composite(com3, img5)
    com5 = Image.alpha_composite(com4, img6)

    # Convert to RGB
    rgbImg = com5.convert('RGB')
    
    # Save to png file
    file_name = str(item["tokenId"]) + ".png"
    rgbImg.save("./outputs/" + file_name)
```
